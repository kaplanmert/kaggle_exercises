```python
from pandas.io.parsers import read_csv

df_2005_2007 = read_csv("/kaggle/input/2000-16-traffic-flow-england-scotland-wales/accidents_2005_to_2007.csv",index_col='Accident_Index')
df_2009_2011 = read_csv("/kaggle/input/2000-16-traffic-flow-england-scotland-wales/accidents_2009_to_2011.csv",index_col='Accident_Index')
df_2012_2014 = read_csv("/kaggle/input/2000-16-traffic-flow-england-scotland-wales/accidents_2012_to_2014.csv",index_col='Accident_Index')
```

```python
df_2005_2007.columns
```

```python
df_2005_2007.head()
```

```python
df = pd.concat([df_2005_2007,df_2009_2011,df_2012_2014])
```

```python
df.info()
```

```python
df.describe().T
```

```python
for col in df.columns:
    df = df[df[col]!=-1]
    
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12,6))
sns.countplot(df['Day_of_Week'],hue=df['Accident_Severity'])
plt.title('Number of accidents each day',)
plt.legend(title="Accident Severity",labels=["Fatal","Serious","Slight"])
df['Day_of_Week'].value_counts()
```

```python
plt.figure(figsize=(13,6))
sns.countplot(df['Speed_limit'],hue=df['Accident_Severity'])
plt.title('The severity of the accident according to the speed limit',size=30)
plt.legend(title="Accident Severity",labels=["Fatal","Serious","Slight"])
```

```python
plt.figure(figsize=(13,6))
sns.countplot(df['Weather_Conditions'],hue=df['Accident_Severity'])
plt.xticks(rotation=70)
plt.title('The severity of the accident according to the weather conditions',size=30)
plt.legend(title="Accident Severity",labels=["Fatal","Serious","Slight"])
```

```python
hours = pd.DataFrame(df["Time"].astype("datetime64"))
hours = pd.DataFrame(hours.Time.dt.strftime('%H').sort_values(),columns=["Time"])
```

```python
plt.figure(figsize=(22,10))
sns.countplot(hours['Time'])
plt.title("Number Of Accidents by Hours")
plt.xlabel("Hours")
plt.ylabel("Number Of Accidents")
```
